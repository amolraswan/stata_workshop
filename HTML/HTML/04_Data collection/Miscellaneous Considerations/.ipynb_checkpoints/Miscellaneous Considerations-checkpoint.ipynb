### 6. Miscellaneous Considerations

Before data collection begins, it is important to ensure that the survey creates and preserves the existence of unique IDs. This means that there must be at least one variable in the data that is unique at the unit level of observation e.g. a household ID, a respondent ID, etc. Ideally, this would be an ID that is automatically generated by the programming (if using digital data collection, this can be done with a simple calculate field). 

If a survey was conducted in a previous round in which unique IDs were first generated, then it is important to consider how best to preserve those IDs in a follow up round of data collection, rather than creating them anew. The consistency in unique IDs across multiple rounds of data collection is what allows us to eventually merge the various rounds together, or compare data accurately across different rounds. To pull in previously created IDs, we can make use of a concept called “preloading”. By pre-populating certain variables’ values into a survey, such as the ID, we can ensure that they get “pulled” in consistently across the different rounds. In SurveyCTO, this feature is called “pulldata” for that very reason. If you are working on a project for which previous data exists, make sure to understand this feature and determine whether it’s necessary for the sake of maintaining unique IDs.
